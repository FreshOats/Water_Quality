I want to RANK the top 3 (highest aka worst) for each parameter

```sql
SELECT 
    *
FROM (
    SELECT  
        station_id, 
        station_type, 
        parameter, 
        result, 
        county_name, 
        DENSE_RANK() OVER(PARTITION BY parameter ORDER BY result DESC) AS Result_Rank
    FROM 
        lab_results
    WHERE result <> 0
) AS lab
WHERE lab.Result_Rank < 4
```

Find the Average Result per parameter per county

```sql
SELECT DISTINCT county_name, parameter, avg_result, max_result
FROM (
    SELECT 
        county_name, 
        parameter, 
        AVG(result) OVER(PARTITION BY parameter, county_name) AS avg_result,
        MAX(result) OVER(PARTITION BY parameter, county_name) AS max_result
    FROM 
        lab_results
    WHERE result > 0
) AS Stats 
ORDER BY parameter, county_name;
```

```sql
SELECT DISTINCT county_name, parameter, avg_result, max_result, rnk
FROM (
    SELECT 
        county_name, 
        parameter, 
        AVG(result) OVER(PARTITION BY parameter, county_name) AS avg_result,
        MAX(result) OVER(PARTITION BY parameter, county_name) AS max_result, 
        RANK() OVER(PARTITION BY parameter ORDER BY result) AS rnk
    FROM 
        lab_results
    WHERE result > 0
) AS Stats 
WHERE rnk < 4
ORDER BY parameter, rnk;
```
