**Starting the analysis of lab results**

Based on the data from the state and federal regulations from California, I can now compare the levels reported on the contaminants to those that were measured. The first issue that I have to consider is the units mismatch.Â 

Many of the units reported in the lab findings are in ug/L, whereas the state report is in mg/L. Since this is off by a factor of 1000, and I only want to look at the values that have this exact mismatch, I'm using a calculated column to compare, dividing the ug value by 1000 to convert to mg/L.

```sql
SELECT  lab.station_id AS Station,  
            lab.parameter AS Contaminant, 
            lab.result AS Result, 
            lab.units AS Micro_units, 
            lab.result / 1000 AS Milli_units, 
            regs.state_max_level AS State_Max, 
            regs.federal_max_level AS Federal_Max, 
            regs.units AS Units
    FROM lab_results lab
        INNER JOIN 
        state_regulations regs
        ON lab.parameter = regs.contaminant
    WHERE   lab.units = 'ug/L' 
            AND 
            regs.units = 'mg/L'
            AND 
            lab.result > 0
```

The next step is to find what factor or percentage of the state and federal maximums are these contaminants.

Since I want to pull directly from the above table's results, I can use a WITH clause to set up a common table expression.

```sql
WITH micrograms_converted_to_milligrams AS 
(   SELECT  lab.station_id AS Station,  
            lab.parameter AS Contaminant, 
            lab.result AS Result, 
            lab.units AS Micro_units, 
            lab.result / 1000 AS Milli_units, 
            regs.state_max_level AS State_Max, 
            regs.federal_max_level AS Federal_Max, 
            regs.units AS Units, 
            lab.county_name AS County
    FROM lab_results lab
        INNER JOIN 
        state_regulations regs
        ON lab.parameter = regs.contaminant
    WHERE   lab.units = 'ug/L' 
            AND 
            regs.units = 'mg/L'
            AND 
            lab.result > 0
)
SELECT  County, 
        Station,
        Contaminant, 
        CAST (100*(Milli_units /  State_Max) AS DECIMAL(18, 3)) AS Percent_of_State_Max, 
        CAST (100*(Milli_units / Federal_Max) AS DECIMAL(18,3)) AS Percent_of_Federal_Max
FROM micrograms_converted_to_milligrams;
```

Now for the fun part. Since it was immediately clear that a few of these were higher than 100% of the state maximum, I want to see which ones and where they are located.

```sql
WITH micrograms_converted_to_milligrams AS 
(   SELECT  lab.station_id AS Station,  
            lab.parameter AS Contaminant, 
            lab.result AS Result, 
            lab.units AS Micro_units, 
            lab.result / 1000 AS Milli_units, 
            regs.state_max_level AS State_Max, 
            regs.federal_max_level AS Federal_Max, 
            regs.units AS Units, 
            lab.county_name AS County
    FROM lab_results lab
        INNER JOIN 
        state_regulations regs
        ON lab.parameter = regs.contaminant
    WHERE   lab.units = 'ug/L' 
            AND 
            regs.units = 'mg/L'
            AND 
            lab.result > 0
), 
percentages_of_converted_contaminants AS
(
    SELECT  County, 
            Station,
            Contaminant, 
            CAST (100*(Milli_units /  State_Max) AS DECIMAL(18, 3)) AS Percent_of_State_Max, 
            CAST (100*(Milli_units / Federal_Max) AS DECIMAL(18,3)) AS Percent_of_Federal_Max
    FROM micrograms_converted_to_milligrams
)
SELECT * 
FROM percentages_of_converted_contaminants
WHERE Percent_of_State_Max >= 100
ORDER BY Percent_of_State_Max DESC;
```
